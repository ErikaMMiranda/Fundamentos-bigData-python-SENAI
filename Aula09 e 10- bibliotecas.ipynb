##### Primeiros passos com Matplotlib

```python
## Importando as bibliotecas

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np



# Criando as arrays (matrizes) para utilizar nos graficos 

x= np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# Criando um grafico com os valores anteriores
plt.plot(x,y)
```

##### MODIFICANDO O TIPO DE MARCADOR


```python
plt.plot (x,y, marker = '*', c = 'red')
```

#### MUDANDO COMO OS PONTOS SÃO LIGADOS E O TIPO DE LINHA


```python
plt.plot (x,y, marker = '*', c = '#DE2686', ds = 'steps',# forma de ligação
           ls = '--') # estilo da linha
```

#### ALTERANDO AS CORES DOS MARCADORES

```python
plt.plot (x,y, marker = '*', c = '#5926DE', ds = 'steps', ls = '--', mec= '#DE2686')
#mec é para mudar a cor do marcador
# as cores foram retiradas do adobe color
```

#### Adicionando outros elementos ao grafico

```python
plt.plot(x,y) #criando o grafico
#adicionado o titulo do eixo X
plt.xlabel('Valores de X', fontsize=12)
#adicionado o titulo do eixo Y
plt.ylabel('Valores de Y', fontsize=12)
# Adicionando titulo do grafico
plt.title ('Título do gráfico', fontsize = 20, c='red') 
# Cor dos rotulos
plt.xticks (fontsize = 8, c = 'blue') 
plt.yticks(fontsize = 8, c = 'purple')
```

##### Aula 10 - MÚLTIPLAS CURVAS NO MECANISMO GRÁFICO   

```python
x = np.linspace(0,2 * np.pi, 100) # esses são os valores do circulo trigonometrico em (np.pi) # gerar 100 numeros entre 0 e 2*pi
y= np.sin(x) # sin é de seno
z= np.cos(x) # cos é de cosseno

plt.plot(x,y, label = 'Função Seno', c='#E68E73')
plt.plot(x,z, label="Função Cosseno", c='#E65931')
plt.plot (x,np.sqrt(x), label="Raiz quadrada", c='#E6B673') # este é o valor da raiz quadrada 
plt.legend(loc='upper left', fontsize= 10) # Adicionar leganda
plt.title('Funções Trigonométricas', color='#EB775B') # adicionar titulo



## Salvar o grafico como imagem  plt.savefig ('caminho da pasta')
plt.savefig("C:\\Users\\FIC\\Documents\\fundamentos-datapython-erika\\Graficos\\funções.jpg")
```

### Gráfico de Pizza

```python
# Exemplo 1 
# Dados para plotar no grafico 

languages= ['Java','Python','PHP','JavaScript','C#','C++']
popularidade= [22.2,17.6,8.8,8,7.7,6.7]
#Explosão das fatias 
explosao = [0.1,0,0,0,0,0] # para destacar o elemento, sendo o primeiro numero o que determina qual o elemento vai ser destacado
# Definido cores
cores=['#EB9BE8','#EBB3A2','#EAA7A2','#EBA2BD','#EBC0A2','#EB9C99']
# Gerando o grafico 
#plt.pie('Primeiro os vetores', 'os rotulos',percentuais, sobreamento= shadow )
#labels = rotulos  de categoria | autopct = rotulos numericos
# %1.1f% > formatando os valores dos rotulos numericos com 1 casa apos a virgula
#definir o angulo de inicio = startangle
plt.pie(popularidade,labels=languages, autopct='%1.1f%%',shadow='true', explode=explosao, colors=cores,startangle=140)
```

```python
dados = pd.read_csv("C:\\Users\\FIC\\Documents\\fundamentos-datapython-erika\\Arquivos\\sea-level.csv")

# Fatiar o data frame - seleciona a partir da linha 6 e todas as colunas
dados = dados.loc[6:,:]
dados.columns = ['Year','CSIRO - Adjusted sea level (inches)','CSIRO - Lower error bound (inches)','CSIRO - Upper error bound (inches)','NOAA - Adjusted sea level (inches)']


# Converter os dados em numericos 
dados['Year'] = dados['Year'].astype(int)
dados['CSIRO - Adjusted sea level (inches)'] = dados['CSIRO - Adjusted sea level (inches)'].astype(float)
dados['CSIRO - Lower error bound (inches)'] = dados['CSIRO - Lower error bound (inches)'].astype(float)
dados['CSIRO - Lower error bound (inches)'] = dados['CSIRO - Lower error bound (inches)'].astype(float)
dados['CSIRO - Upper error bound (inches)'] = dados['CSIRO - Upper error bound (inches)'].astype(float)
dados['NOAA - Adjusted sea level (inches)']= dados['NOAA - Adjusted sea level (inches)'].astype(float)
dados.info()


```

```python
anos = dados['Year'].to_list()
nivel = dados['CSIRO - Adjusted sea level (inches)'].to_list()

sns.scatterplot(x=anos,y=nivel)
plt.xlabel('Ano')
plt.ylabel('Nivel do mar (polegadas)')
plt.title('Aumento do nivel do mar')
plt.show
```
