## Etapa 4 - Criar e testar o modelo    

```python
# Criação da variavel dependente (y) > variavel resposta
y = dados['Revenue']

# Criação da variavel independente (x) > entrada
X = dados[['Temperature']] 

# separar os dados para teste e treino 

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25) 
```

### Etapa 5 - treinado o modelo 

```python
# Criação do modelo 
modelo = LinearRegression(fit_intercept=True)

# treinamento do modelo 
modelo.fit(X_train,y_train)

# exibição dos coeficientes 

print('Coeficiente linear (b):', modelo.intercept_)
print('Coeficiente angular (a):',modelo.coef_)
```

### Etapa 6 - Testando o modelo 

```python
y_predict = modelo.predict(X_test)
y_predict  
```
