##### Exercício 02 - Obtenha o lucro total de todos os meses e mostre o gráfico de linhas com as seguintes propriedades de estilo.
##### O gráfico de linha gerado deve incluir as seguintes propriedades de estilo:

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os 
```

```python
df = pd.read_csv("C:\\Users\\FIC\\Documents\\fundamentos-datapython-erika\\Arquivos\\company_sales_data.csv")
df.head()
```

```python
listaLucro= df['total_profit']
listaMes= df['month_number']

plt.plot(
    listaMes,
    listaLucro, 
    label = 'Lucro Mensal do Ano Anterior', 
    ls= '--', 
    c = 'red',marker = 'o',
    markerfacecolor= 'k',#remove a borda do marcador
    mec = 'black',
    linewidth = 3 #espesura da linha 
)

plt.xlabel('Numero do mes', fontsize = 12)
plt.ylabel('Número de unidades vendidas', fontsize = 12)
plt.title ('Company Sales data of last year', fontsize = 15, c='black') 
plt.yticks([100000,200000,300000,400000,500000])# ajustando os intervalos do eixo y 
plt.xticks(listaMes)# Ajustando os rotulos do eixo X
plt.legend(loc='lower right') # LOC= permite a mudança da posição da legenda
```

#### Exercício 03 – Leia todos os dados de vendas de produtos e mostre-os usando um gráfico multilinha. Exiba o número de unidades vendidas por mês para cada produto usando gráficos multilinhas. (ou seja, faça a plotagem das linhas separadas para cada produto).

```python
# Primeira etapa : Criar listas com base nas colunas do dataframe

listaMeses = df['month_number'].to_list()
faceCreamSalesData = df['facecream'].to_list()
facewasSalesData = df['facewash'].to_list()
toothpasteSalesData = df['toothpaste'].to_list()
bathingsoapSalesData= df['bathingsoap'].to_list()
shampooSalesData= df['shampoo'].to_list()
moisturizerSalesData= df['moisturizer'].to_list()
#Gera um grafico de linhas para plotar cada variavael descrita acima
#Tipo de marcador = o | lidewidth = 3

plt.plot(listaMeses,faceCreamSalesData, label ="Face Cream", marker = 'o', linewidth = 3)
plt.plot (listaMeses,facewasSalesData, label ="Face Wash", marker = 'o', linewidth= 3)
plt.plot (listaMeses,toothpasteSalesData, label ="Tooth Paste", marker = 'o', linewidth= 3)
plt.plot (listaMeses,bathingsoapSalesData, label ="Bathing Soap", marker = 'o', linewidth= 3)
plt.plot (listaMeses,shampooSalesData, label ="Shampoo", marker = 'o', linewidth= 3)
plt.plot (listaMeses,moisturizerSalesData, label ="Moisturizer", marker = 'o', linewidth= 3)
plt.legend(loc='upper left')
plt.yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
plt.xlabel('Numero do mes', fontsize = 10)
plt.ylabel('Unidades Vendidas', fontsize = 10)
plt.title('Dados de venda')
```

##### Exercício 04 - Leia os dados de vendas de pasta de dente de cada mês e mostre-os usando um gráfico de dispersão.

```python
import seaborn as sns   
listaMeses = df['month_number'].to_list()
toothpasteSalesData = df['toothpaste'].to_list()

#criar o grafico 
sns.scatterplot(x=listaMeses,y=toothpasteSalesData)
plt.xlabel('Meses', c= 'blue')
plt.ylabel('Numero de unid. vendidas', c= 'blue')
plt.title('Vendas de Creme dental por mês', fontsize=15, c='purple')
plt.grid(True, linewidth= 0.5, linestyle= '--', c='#D9F5BA') # Adiconar grade (plt.grid)
```
